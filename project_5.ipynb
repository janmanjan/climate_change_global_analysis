```python
#reading in all our north pole ice melting data
ice_1 = pd.read_csv('/content/N_01_extent_v3.0.csv', index_col=None)
ice_2 = pd.read_csv('/content/N_02_extent_v3.0.csv', index_col=None)
ice_3 = pd.read_csv('/content/N_03_extent_v3.0.csv', index_col=None)
ice_4 = pd.read_csv('/content/N_04_extent_v3.0.csv', index_col=None)
ice_5 = pd.read_csv('/content/N_05_extent_v3.0.csv', index_col=None)
ice_6 = pd.read_csv('/content/N_06_extent_v3.0.csv', index_col=None)
ice_7 = pd.read_csv('/content/N_07_extent_v3.0.csv', index_col=None)
ice_8 = pd.read_csv('/content/N_08_extent_v3.0.csv', index_col=None)
ice_9 = pd.read_csv('/content/N_09_extent_v3.0.csv', index_col=None)
ice_10 = pd.read_csv('/content/N_10_extent_v3.0.csv', index_col=None)
ice_11 = pd.read_csv('/content/N_11_extent_v3.0.csv', index_col=None)
ice_12 = pd.read_csv('/content/N_12_extent_v3.0.csv', index_col=None)

ice_1.head()
```

```python
#combining each month of the north pole data to a csv
north_ice = pd.concat(
                      [ice_1, ice_2, ice_3, ice_4, ice_5, ice_6, ice_7, ice_8, ice_9, ice_10, ice_11, ice_12]
                      , axis=1)
#saving cleaned north pole data frame to a csv
north_ice.to_csv('./cleaned_data/north_ice.csv', index=False)
```
